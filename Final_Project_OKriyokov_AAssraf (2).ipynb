## 5. Web Scraping Technique
<a id = "5"></a>

The second dataset for this project has been explored through scraping the GVA website, https://www.gunviolencearchive.org. 
We scraped the data which was most relevant to our research questions. 
We first examined the data on the GVA website by inspecting the html code of this page- https://www.gunviolencearchive.org/reports, due to the fact that the data which we needed started from this page. We further proceeded by finding out different tags like, 'table'- the data which we needed was in table. 

Then we further looked for 'href' tag, it had the link to the next page, which helped in sliding over the pages untill the end. 
Finally, we exported the data we found into CSV file. 
We did not encounter any missing values while scrapping the GVA website. 


### 5.1. Web Scraping Procedure
<a id = "5.1."></a>

```python
def scrape_data(soup):
    table = soup.find("table")
    for row in table.find_all('tr')[1:]:
        col = row.find_all('td')
        date = col[0].string
        state= col[1].string
        city = col[2].string
        killed = col[4].string
        injured = col[5].string

        writer.writerow({str(headings[0]): str(date), str(headings[1]): str(state), str(headings[2]): str(city),
                        str(headings[3]): str(killed),str(headings[4]): str(injured)})


file_path = 'C:/Users/Olga Kriyokov/Dataset+Final/test.csv'

csv1 = open(file_path, mode='w',newline = '')
headings = ['Date','State','City_country','n_killed','n_injured']
writer = csv.DictWriter(csv1, fieldnames=headings)
writer.writeheader()

for i in range(5):
    years=[2014,2015,'year=2016','year=2017','year=2018']
    base_url = 'https://www.gunviolencearchive.org/reports'
    next_url = '/mass-shootings/'+str(years[i])+'?sort=asc&order=Incident%20Date'

    if i >=2:
        next_url = '/mass-shooting?'+str(years[i])

    while next_url:
        next_page = requests.get(base_url+next_url)
        print(base_url+next_url)
        base_url = 'https://www.gunviolencearchive.org'
        
        c = next_page.content
        soup = BeautifulSoup(c,'lxml')
        scrape_data(soup)
        next_url = soup.find('a', {'title': 'Go to next page'})
        if next_url:
            next_url = next_url.get('href')
            print(next_url)
csv1.close()
```

### 7.1: How have gun sales changed between the years of 2014- 2018?
<a id = "7.1"></a>

To be able to answer this question, we will use a time series charts for each year which illustrate data points at successive interval of time. In the below charts we will display the fluctuation of gun sales during the different years.

```python
def time_series_charts(num):
    while(num<=2018):
        aggregate = {'state':'count','totals':'sum'}
        x = data_1[data_1['year'] == num].groupby('date').agg(aggregate).reset_index()
        data = [go.Scatter(x=x.date,y=x.totals,mode = "lines")]
        layout = dict(height=350, title = 'Fluctuation of Gun sales during the year-'+str(num), xaxis= dict(title='Month-Year', ticklen= 2))
        num = num+1
        fig = dict(data = data, layout = layout)
        py.offline.iplot(fig)
    
    a = data_1['year'] > 2013 
    b = data_1['year'] < 2019
    c = a & b
    x = data_1[c].groupby('year').agg(aggregate).reset_index()
    data = [go.Scatter(x=x.year,y=x.totals,mode = "lines")]
    layout = dict(height=400, title = 'Fluctuation of Gun sales during the years-2014-2018', xaxis= dict(title='Year', ticklen= 2))
    fig = dict(data = data, layout = layout)
    py.offline.iplot(fig)

time_series_charts(2014)
```
