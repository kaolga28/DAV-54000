# FIRST (DATASET TEST RESULTS)

```python
data=pd.read_excel('Test results.xlsx')
```

```python
print (data.head(2))
```

### Data that is read, has too much junk

```python
#Filtering data
cols_to_keep=['Name', 'Test 1', 'Test 2', 'Test 3', 'Test 4']
filtered_data=data[cols_to_keep]
print (filtered_data.head(2))
```

```python
#filtered data has all NAN rows so delete where every column entry is null
filtered_data=filtered_data.dropna(how='all') 
#Not dropping single NaN because they represent absent from the specific test
#Final filtered data
print (filtered_data)
```

### Questions

```python
#1. Which student had the highest average test grades?
filtered_data['total']=filtered_data.sum(axis=1)
print ("highest average test grades: ",filtered_data['Name'].loc[filtered_data[['total']].idxmax()] )
#2. Which student had the greatest range of test grades?
#tweaking nans to get max
filtered_data.fillna(-100, inplace=True)
filtered_data['Max'] = filtered_data[['Test 1', 'Test 2', 'Test 3', 'Test 4']].max(axis=1)
#tweaking nans to get min
filtered_data.replace(-100,101,inplace=True)
filtered_data['Min'] = filtered_data[['Test 1', 'Test 2', 'Test 3', 'Test 4']].min(axis=1)
#getting nans back
filtered_data.replace(101,np.nan,inplace=True)
filtered_data['Diff']=filtered_data['Max']-filtered_data['Min']
print (filtered_data)
print ("greatest range of test grades: ",filtered_data['Name'].loc[filtered_data[['Diff']].idxmax()] )
#3. Which test had the highest grades?
temp=filtered_data[['Test 1', 'Test 2', 'Test 3', 'Test 4']].mean(axis=0)
print ("all:\n",temp)
print ("max: ",max(temp), "\nso answer is Test 4")
```

# THIRD (DATASET Fifa)

```python
data=pd.read_excel('fifa.xlsx')
```

```python
data.head(20)
```

## DATA CLEANING

```python
#CLEANING COLUMN BY COLUMN
#AGE
#Age is always supposed to be non negative.
data = data.drop(data[data['Age']<0].index)
#one record with age in negative dropped
#OVERALL
#skills are rated from 0 to 100 there must be no nan value in overall columns for a clean dataset
data = data.drop(data[data['Age']<0].index)
data=data[np.isfinite(data['Overall'])]
#VALUE 
#value of player must be in positive
data = data.drop(data[data['Value (M€)']<0].index)
#Preferred Foot
# preffered foot must only be right or left
data1 = data[data['Preferred Foot']=="Right"]
data2 = data[data['Preferred Foot']=="Left"]
data=data1.append(data2)
#INTERNATIONAL REPUTATION
#MUST BE OUT OF 5 AS PER SPECIFIED
data = data.drop(data[data['International Reputation (out of 5)']>5].index)
#WEAK FOOT
#MUST BE OUT OF 5 AS PER SPECIFIED
data = data.drop(data[data['Weak Foot(out of 5)']>5].index)
data=data.reset_index()
print ("CLEANED DATA\n")
print (data)
```

# Questions

```python
#BEST RATING PLAYERS
print ("best rated player")
print (data.loc[data['Overall'].idxmax()])
#CLUBS WITH MOST PLAYERS
print ("Club with most players")
data['Club'].value_counts()[data['Club'].value_counts() == data['Club'].value_counts().max()]
```
